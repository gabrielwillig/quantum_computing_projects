```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
from qiskit_algorithms import HamiltonianPhaseEstimation
```

```python
circuit = QuantumCircuit(2)

circuit.h(0)
circuit.t(0)
circuit.cx(0, 1)

display(circuit.draw())
```

```python
statistics = Statevector.from_label('00').evolve(circuit).sample_counts(1000)
display(statistics)
```

```python
plot_histogram(statistics)
```

```python
circuit_2 = QuantumCircuit(3)

circuit_2.h(0)
circuit_2.h(1)
circuit_2.ccx(0, 1, 2)

display(circuit_2.draw())
```

```python
statistics_2 = Statevector.from_label('000').evolve(circuit_2).sample_counts(1000)
display(statistics_2)
```

```python
plot_histogram(statistics_2)
```
