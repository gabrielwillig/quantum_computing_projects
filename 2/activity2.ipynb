# 2. Construa um circuito quântico e aplique sobre ele a porta de Hadamard
A porta de Hadamard é dada pelo operador $H = (1 / √2) * ( |0><0| + |0><1| + |1><0| - |1><1| )$  
para construir, no Qiskit, um circuito quântico que recebe o estado |0>, aplica sobre ele a porta de Hadamard, e depois faz uma medição do qubit resultante. Realize experimentos para plotar um histograma sobre os resultados das medições. 

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
```

```python
circuito_quantico = QuantumCircuit(1)
# Aplicando a porta de Hadamard
circuito_quantico.h(0)
circuito_quantico.draw()
```

```python
ket0 = Statevector([1, 0])
v = ket0.evolve(circuito_quantico)
v.draw("text")
```

```python
estatísticas = v.sample_counts(4000)
plot_histogram(estatísticas)
```
