# 1. Re-escreva o estado
((1 + 2i) / 3)|0> - (2/3)|1>
de modo que a amplitude de |0> seja um número real.
## Solução:
### Amplitude de |0>  
$\frac{1}{3} + \frac{2i}{3}$
### Fator Global
$e^{-i\theta}$  
Onde, $\theta = tan^{-1}(\frac{\frac{2}{3}}{\frac{1}{3}}) = tan^{-1}(2)$  
### Multiplicamos o estado pelo fator global
$(\frac{1+2i}{3}|0> - \frac{2}{3}|1>) e^{-i\theta}$  
#### Resulta em:
$\frac{\sqrt{5}}{3} |0>  - \frac{2}{3}e^{-i\theta} |1>$


```python
from qiskit.quantum_info import Statevector
from numpy import sqrt
from cmath import exp, cos, sin
from qiskit.visualization import plot_histogram
```

```python
v = Statevector([(1 + 2.0j) / 3, -2 / 3])

display(v.is_valid())
```

```python
u = Statevector([sqrt(5)/3, -(2/3)*exp(-1.0j*(cos(2)/sin(2)))])

display(u.is_valid())
```

```python
v.draw("latex")
```

```python
u.draw("latex")
```

```python
statistics = v.sample_counts(1000)
display(statistics)
plot_histogram(statistics)
```

```python
statistics_u = u.sample_counts(1000)
display(statistics_u)
plot_histogram(statistics_u)
```
